```python
passage = """It was a busy Monday morning, and I woke up feeling tired.
The alarm had gone off at 6 AM, but I hit the snooze button several times before finally getting out of bed. The sun was barely rising, and the cool morning air filled the room.
I headed to the kitchen to make a cup of coffee. As I stood by the window, sipping my coffee, I thought about the long day ahead. I had several meetings, a few reports to finish, and a couple of calls to make.

After a quick shower, I dressed in my usual office clothes and grabbed my bag.
On my way out, I remembered to check if I had my wallet and keys.
"Where are my keys?" I muttered to myself as I searched through the bag. After a few moments, I found them tucked in a small pocket.
Relieved, I locked the door and headed to the bus stop. The street was busy with people rushing to work, and cars were honking in the distance.

I arrived at the office just in time for the first meeting. My boss greeted me with a nod as I walked into the conference room.
The meeting dragged on for nearly two hours, and by the end of it, I was exhausted. There was a lot to catch up on, and deadlines were fast approaching.
My colleagues and I exchanged notes and planned out the week.
"Let’s get this done by Friday," said my boss, as he handed out the assignments.

During lunch, I decided to take a short walk to clear my mind.
The weather was nice, and the fresh air helped me feel more focused.
 I stopped by a small café to grab a sandwich. As I sat down, I checked my phone and saw a few messages from friends, reminding me about our dinner plans for the weekend.
 I replied quickly and got back to the office.

The afternoon was busy. I spent hours working on the reports, barely looking up from my desk. Time flew by, and before I knew it, it was already 6 PM.
I packed my things and headed home, feeling tired but satisfied with the work I had done. On the way home, I decided to pick up some groceries for the week. The store was crowded, but I quickly gathered what I needed—milk, bread, vegetables, and a few snacks.

When I got home, I put the groceries away and sat down on the couch.
I scrolled through my phone, checking social media and catching up on the day’s news.
The evening felt peaceful. After a simple dinner of pasta and salad, I decided to watch a movie.
I turned on my favorite streaming service and browsed through the options.
After some time, I picked a comedy and settled in for the night.

As the movie ended, I realized how tired I was.
It had been a long day, and my eyes were starting to close.
I turned off the TV and headed to bed. As I lay down, I thought about the next day and all the things I still had to do.
But for now, I was content to relax. The sound of the city outside was calming, and soon, I drifted off to sleep, ready to do it all over again the next day."""
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([passage])
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
input_sequences = []
for sentence in passage.split('\n'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]

  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1])
```

```python
input_sequences
```

```python
max_len = max([len(x) for x in input_sequences])
max_len
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sequences, maxlen = max_len, padding='pre')
padded_input_sequences
```

```python
x = padded_input_sequences[:,:-1]
```

```python
y= padded_input_sequences[:,-1]
```

```python
x.shape
```

```python
y.shape
```

```python
x
```

```python
y
```

```python
from tensorflow.keras.utils import to_categorical
y = to_categorical(y,num_classes=260)
```

```python
y.shape
```

```python
y
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense, Input

# Define the model
model = Sequential()
model.add(Input(shape=(56,)))
model.add(Embedding(input_dim=260, output_dim=100))
model.add(LSTM(150, return_sequences=True))
model.add(LSTM(150))
model.add(Dense(260, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy',metrics=['accuracy'])
model.summary()
```

```python
model.fit(x,y,epochs=30)
```

```python
import time
import numpy as np
text = "It was a busy Monday"

for i in range(10):
  # tokenize
  token_text = tokenizer.texts_to_sequences([text])[0]
  # padding
  padded_token_text = pad_sequences([token_text], maxlen=48, padding='pre')
  # predict
  pos = np.argmax(model.predict(padded_token_text))

  for word,index in tokenizer.word_index.items():
    if index == pos:
      text = text + " " + word
      print(text)
      time.sleep(2)
```
